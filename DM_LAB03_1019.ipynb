```python
ypoints = np.array([3,8,1,20,5,7])
plt.plot(ypoints, marker = 's')
plt.show()
```

```python
ypoints = np.array([3,8,1,20,5,7])
plt.plot(ypoints, 's:c')
plt.show()
```

```python
plt.plot(ypoints,ls='-.',marker='s',ms=20,mec='royalblue',mfc='y',linewidth='3')
plt.show()
```
