```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
```

```python
#sample dataset (3D data)

data = {
'Feature1': [2, 4, 6, 8, 10],
'Feature2': [1, 3, 5, 7, 9],
'Feature3': [5, 10, 15, 20, 25] }
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Standardize the Data (PCA requires normalization)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# 3D visualization (before PCA)

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Feature2')
ax1.set_zlabel('Feature3')
ax1.set_title('Before PCA (3D)')
ax1.legend()
ax1.scatter(df['Feature1'], df['Feature2'], df['Feature3'], color='blue', label='Original Data')
```

```python
# Apply PCA (Reduce from 3D to 2D)
pca = PCA(n_components=2)
```

```python
#Convert to DataFrame
principalComponents = pca.fit_transform(scaled_data)
```
